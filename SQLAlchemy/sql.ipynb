```python
from sqlalchemy import create_engine,text 
```

```python
engine = create_engine('sqlite:///mydatabase.db',echo=True)
```

```python
conn = engine.connect()
```

```python
conn.execute(text("create table if not exists users (id integer primary key autoincrement, name text, email text)"))
conn.commit()
```

```python
#mysql+pymysql://username:password@host:port/database
engine1 = create_engine("mysql+pymysql://root:1234@localhost:3306/okay",echo=True)
```

```python
conn1=engine1.connect()
conn1.begin() # begin is used to manitain atomicity as error happens it rollback automaically and if no error then commit automatically
result = conn1.execute(text("show tables"))
for row in result:
    print(row)
conn1.close()

```

```python
engine = create_engine('mysql+pymysql://root:1234@localhost',echo=True)
conn=engine.connect()
conn.begin()
```

```python
conn.execute(text("create database sqlalchemy"))
conn.execute(text("use sqlalchemy"))
conn.execute(text("create table user (id int primary key , name varchar(50) , age int )"))
conn.execute(text("insert into user values (1,'shreeram', 25)"))
conn.close()

```
