# Series 

```python
import pandas as pd
```

using list

```python
data = [10,20,30,40,50,60,70,80,90]
df = pd.Series(data,index=range(1,10))
print(df)
```

```python
data = ['Shreeram','Shree','Ram','Riya','Sushma','John']
df = pd.Series(data,index=range(1,7),dtype=object)
print(df)
```

using dictionary

```python
data = {'name':'shreeram','age':27,'gender':'male','salary':10000}
df = pd.Series(data,dtype=object)
print(df)
```
