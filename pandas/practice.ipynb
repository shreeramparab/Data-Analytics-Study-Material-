````python
import pandas as pd
import numpy as np
````

````python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Hank',
             'Ivy', 'Jack', 'Karen', 'Leo', 'Mia', 'Nate', 'Olivia', 'Paul'],
    'Department': ['HR', 'Tech', 'Tech', 'HR', 'Finance', 'Finance', 'Tech', 'HR',
                   'Tech', 'Finance', 'HR', 'Tech', 'Finance', 'HR', 'Tech', 'Finance'],
    'Salary': [55000, 82000, 91000, 60000, 73000, 68000, 95000, 52000,
               87000, 71000, 58000, 99000, 65000, 53000, 78000, 70000],
    'Years': [3, 5, 7, 2, 4, 6, 8, 1, 5, 3, 2, 9, 4, 1, 6, 3],
    'Rating': [4.1, 3.8, 4.5, np.nan, 4.0, 3.5, 4.7, np.nan,
               4.2, 3.9, np.nan, 4.8, 3.6, np.nan, 4.3, 3.7],
    'City': ['NYC', 'LA', 'NYC', 'Chicago', 'LA', 'NYC', 'Chicago', 'LA',
             'NYC', 'Chicago', 'LA', 'NYC', 'Chicago', 'LA', 'NYC', 'Chicago'],
    'Gender': ['F', 'M', 'M', 'F', 'F', 'M', 'F', 'M',
               'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M']
}
df = pd.DataFrame(data)
````

````python
'''🟡 Section 1 — Basics & Inspection (Q1–Q6)
Q1. Display the first 5 rows, shape, column names, and data types of df — all in one go.
Q2. How many employees are in each Department? Show the result sorted in descending order.
Q3. Find the total number of missing values per column.
Q4. Get all unique cities in the dataset and the count of employees in each.
Q5. Find the highest and lowest salary in the entire dataset.
Q6. Display basic statistics (mean, std, min, max) for only the numeric columns.'''
print(df.head(5),df.shape,df.columns.tolist(),df.dtypes) #1
print(df['Department'].value_counts())#2
print(df.isnull().sum().sum())#2
print(df['City'].unique().tolist())#4
print(df.select_dtypes(include=['number']).agg(['mean','sum','std','min','max']))#5
````

````python
'''
🟠 Section 2 — Filtering & Indexing (Q7–Q13)
Q7. Filter employees from the Tech department earning more than 85,000.
Q8. Get all employees from NYC or Chicago who have a Rating above 4.0.
Q9. Select employees whose name starts with the letter 'A', 'B', or 'C'.
Q10. Find employees who are in Finance and have worked for less than 4 years.
Q11. Use .loc[] to select rows where Years is between 3 and 6 (inclusive), and show only Name, Department, Salary.
Q12. Use .iloc[] to select the last 5 rows and the first 3 columns.
Q13. Find all employees whose Salary is above the overall mean salary.'''
print(df[(df['Department']=='Tech')&(df['Salary']>85000)])#1
print(df[df['City'].isin(['NYC','Chicago'])&(df['Rating']>4.0)])#2
print(df[df['Name'].str.startswith(('A','B','C'))])#3
print(df[(df['Department'] == 'Finance') & (df['Years'] < 4)])#4
print(df.loc[df['Years'].between(3,6),['Name','Department','Salary',]])#5
print(df.iloc[-5:,:3])#6
print(df.loc[df['Salary']>df['Salary'].mean()])#7
````

````python
'''🔵 Section 3 — Cleaning & Transformation (Q14–Q20)
Q14. Fill missing Rating values with the median rating (without modifying the original df).
Q15. Drop all rows where Rating is missing.
Q16. Create a new column "Experience_Level" with values:
"Junior" → Years < 3
"Mid" → 3 ≤ Years ≤ 6
"Senior" → Years > 6
Q17. Rename the columns Salary → Annual_Salary and Years → Experience_Years.
Q18. Convert the Gender column from 'M'/'F' to 'Male'/'Female' using .map().
Q19. Create a "Salary_Band" column using pd.cut() with 3 equal-width bands labeled "Low", "Medium", "High".
Q20. Reset the index of df after dropping rows where Rating is null. Make sure the old index is not kept as a column.'''
df['Rating'].fillna(df['Rating'].median())#1
df.dropna(subset=['Rating'])#2
data=np.select([df['Years']<3,df['Years']>6],['Junior','Senior'],default='Mid')#3
df['Experience_Level'] = data
df.rename(columns={'Salary':'Annual_Salary','Years':'Experience_Years'},inplace=True) #4
df['Gender'] = df['Gender'].map({'M': 'Male', 'F': 'Female'}) #5
df['Salary_Band'] = pd.cut(df['Annual_Salary'],bins=3,labels=['Low', 'Medium', 'High'])#6 
'''Range = Max - Min = 120,000 - 40,000 = 80,000

### Step 2 — It divides into 3 **equal-width** bins
```
Bin 1 "Low"    → 40,000  to  66,666
Bin 2 "Medium" → 66,667  to  93,333
Bin 3 "High"   → 93,334  to 120,000
```
> 💡 Each bin covers the **same salary range (≈26,666)** — that's what *equal-width* means.

### Step 3 — It assigns a label to every employee
```
Employee A → Salary 45,000  → "Low"
Employee B → Salary 75,000  → "Medium"
Employee C → Salary 110,000 → "High"'''
df = df.dropna(subset=['Rating']).reset_index(drop=True) #7
````

````python
'''🔴 Section 4 — GroupBy & Aggregation (Q21–Q27)
Q21. Find the mean salary grouped by `Department`.
Q22. For each `Department`, find the total salary expenditure and number of employees.
Q23. Group by `Department` and `City`, and find the average rating for each combination.
Q24. Find the top earner (Name + Salary) in each department. (Hint: `idxmax`)
Q25. How many `Male` and `Female` employees are there in each `Department`?
Q26. Use `.agg()` to compute min, max, and mean of `Salary` grouped by `Department`.
Q27. Find departments where the average salary is above 75,000.'''
df.groupby('Department').agg({'Salary': 'sum'})
df.groupby('Department').agg({'Salary': ['sum','count']})
df.groupby(['Department','City'])['Rating'].mean()
df.groupby(['Department','Gender']).size()
df.groupby(['Department']).agg({'Salary': ['mean','min','max']})
df.groupby(['Department'])['Salary'].mean()[lambda x : x>75000]
````

````python
reviews = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Eve', 'Grace', 'Leo'],
    'Review': ['Good', 'Average', 'Excellent', 'Good', 'Excellent', 'Good']
})
dept_budget = pd.DataFrame({
    'Department': ['HR', 'Tech', 'Finance'],
    'Budget': [200000, 500000, 350000]
})
````

````python
'''Q33. Merge df with reviews so all employees are kept and those without a review show NaN.
Q34. Merge df with reviews and keep only employees who have a review (inner join).
Q35. Merge df with dept_budget on Department so each employee also shows their department's budget.
Q36. After merging df with dept_budget, create a column "Salary_to_Budget_Ratio" = Salary / Budget.
'''
pd.merge(df,reviews,on='Name',how='left')
pd.merge(df,reviews,on='Name',how='inner')
pd.merge(df,dept_budget,on='Department')
merged = pd.merge(df,dept_budget,on='Department').assign(Salary_to_Budget_Ratio =lambda x : x['Salary']/x['Budget'])
merged
````
