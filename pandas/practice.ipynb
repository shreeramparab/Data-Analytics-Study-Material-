```python
import pandas as pd
```

```python
data = pd.DataFrame({
                    'id':[1,2,3,4,5],
                    'name': ['Alice', 'Bob', 'Charlie', 'Dennis', 'Edith'],
                    'age': [20, 25, 30, 35, 40],
                    'height': [1.76, 1.67, 1.58, 1.43, 1.50],
                    'weight': [69, 62, 58, 54, 50]})
# first method 
data.index = range(1,len(data)+1)
data.index.name = 'id'
# second method 
data['Sr.no'] = [i for i in range(1,len(data)+1)]
data.index = data['Sr.no']
# third method 
data.set_index('Sr.no', inplace=True)
# drop
data.reset_index(inplace=True,drop=True)
print(data)
```

```python
now = pd.Timestamp.now()
df = pd.read_csv('names.csv',index_col=0,header=0,parse_dates=['Dob'],dayfirst=True) # pandas has default read date format i.e MM-DD-YYYY
#df.drop('Age by Months',axis=1,inplace=True)
col = list(df.columns)
df.drop('Age',axis=1,inplace=True)
df.insert(col.index('Dob'),'Age',((now.year - df["Dob"].dt.year)))
print(df)
```
