```python
import pandas as pd
import numpy as np
```

#group by 

```python
df = pd.read_csv('employee.csv',header=0,index_col=0,skipinitialspace=True)
print(df)
```

```python
col = list(df.columns)
#df.drop('dept',axis = 1,inplace=True)
df.insert(col.index('gender')+1,'dept',np.where(df['salary']==12000,'IT',np.where(df['salary']>12000,'Finance','HR')))
print(df)
```

```python
print(df.groupby('dept')['salary'].sum())
print(df.groupby('dept')['salary'].mean())
print(df.groupby('dept')['salary'].count())
print(df.groupby('dept')['salary'].min())
print(df.groupby('dept')['salary'].max())
print(df.groupby('dept').agg({'salary':{'sum','mean','count','min','max'}}))
print(df.groupby('dept').agg({'salary':'sum','bonus':'sum','age':'mean'}).astype(int))
print(df.groupby('dept')['salary'].agg(['sum','mean','count','min','max']).astype(int))
print(df.groupby('dept')['salary'].agg(avg_sal=('mean'),total_sal='sum',total_emp=('count')).astype(int))
```
