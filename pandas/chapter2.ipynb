```python
import pandas as pd
```

```python
df = pd.read_csv('names.csv',header=0,index_col=0)
print(df)
col = list(df.columns)
df.insert(col.index('salary')+1,'bonus',(df['salary']*0.1).astype(int))
print(df)
df.to_csv('names1.csv',index=True,header=True)
col = list(df.columns)
```

```python
df = pd.read_csv('employee.csv',header=0,index_col=0)
print(df)
col = list(df.columns)
df.loc[df['name']=='Shree','gender'] = 'Male'
df.update(df)
print(df)
df.insert(col.index('salary')+1,'bonus',df['salary']*0.1)
print(df)
df.to_csv('employee.csv',index=True,header=True)
```

```python
df = pd.read_csv('employee.csv',header=0,index_col=0)
new_df = pd.DataFrame([{'name':'Shreeram','age':27,'gender':'Male','salary':11237,'bonus':1123}],index=[len(df)+1])
df = pd.concat([df,new_df])
print(df)
df.to_csv('employee.csv',index=True,header=True)
```

```python
load = 2
df = pd.read_csv('employee.csv',header=0,index_col=0,chunksize=load)
for i in df:
    i['salary'] = i['salary'] * 1.10
    print(i)
    i.to_csv('employee3.csv',mode='a',index=True,header=False)
```
