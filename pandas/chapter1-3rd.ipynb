```python
df = pd.read_csv('employee.csv',header=0,index_col=0)
print(df)
col = list(df.columns)
df.loc[df['name']=='Shree','gender'] = 'Male'
df.update(df)
print(df)
df.insert(col.index('salary')+1,'bonus',df['salary']*0.1)
print(df)
df.to_csv('employee.csv',index=True,header=True)
```
