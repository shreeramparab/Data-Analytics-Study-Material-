```python
df = pd.read_csv('employee.csv',header=0,index_col=0)
print(df)
engine = create_engine('mysql+pymysql://root:1234@localhost/okay')
conn = engine.connect()
df.to_sql('employee',conn,if_exists='replace',index=True)
print(df)
```

```python
col = list(df.columns)
city = ['New York','London','Paris','Tokyo','Sydney','Mumbai']
df.drop('city',axis=1,inplace=True)
df.insert(col.index('age')+1,'city',city)
df.to_csv('employee.csv',index=True,header=True)
df.to_sql('employee',conn,if_exists='replace',index=True)
result = conn.execute(text("select * from employee"))
print(df)
for row in result :
    print(row)
```

```python
con = [df['salary'] > 12000 ,df['salary' ]== 12000,df['salary' ]< 12000]
hike = [1.40,1.30,1.20]
df.drop('hike',axis=1,inplace=True)
df.insert(col.index('salary')+1,'hike',0)
df['hike'] = np.select(con,hike)
df.to_csv('employee.csv',index=True,header=True)
df.to_sql('employee',conn,if_exists='replace',index=True)
```
