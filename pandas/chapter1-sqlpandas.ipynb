```python
df= pd.read_csv('employee.csv',header=0,index_col=0)
col = list(df.columns)
df.insert(col.index('gender')+1,'working_hours',0)
new_df = pd.DataFrame({
    'id':range(1,len(df)+1), 
    'working_hours':np.random.randint(5,10,size=len(df))
})
new_df.set_index('id',inplace=True)
```

```python
df.update(new_df)
print(df)
```

```python
df = df.merge(new_df, how='left', on='id')
print(df)
```

```python
avg = df['bonus'].mean()
df['bonus'] = avg.astype(int)
print(df)
```

```python
gender_bonus = df.groupby('gender')['bonus'].transform('mean')
print(df)
df['bonus'] += gender_bonus
print(df)
```
