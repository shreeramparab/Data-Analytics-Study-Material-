```python
df = pd.read_csv('employee.csv',header=0,index_col=0)
print(df)
engine = create_engine('mysql+pymysql://root:1234@localhost/okay',echo=True)
conn = engine.connect()
df.to_sql('employee',conn,if_exists='replace',index=True)
```

```python
result = conn.execute(text("select * from employee"))
for r in result :
    print(r)
```
