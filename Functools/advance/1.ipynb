```python
import functools as f
```

```python
num = [1,2,3,4]
result = f.reduce(lambda x , y : x+y,num)
print(result)
```

```python
def power(base,exp) :
    return base**exp
square = f.partial(power,exp=2) 
print(square(2))
# used to freeze some portion of the function and pass the remaining portion as a value
# here we can use the same function multiple times with the fixed value 
 #here exp = 2 is fixed as we converted the sqaure variable or the object 
 #into the partial fucntion using the f.partial function so we just need to pass the one value """
 
```

```python
#partial function but for class
class myclass:
    def greet (self,msg):
        print(msg)
    say_hello = f.partialmethod(greet,'hello')
obj = myclass()
obj.say_hello()
```
