lru_cache

📌 Used for:
1) Expensive recursive calls
2) API calls
3) Computation-heavy functions

```python
@f.lru_cache(maxsize=None)
def fun_cache(x):
    if x < 2:
       return x
    else:
        a , b =0,1
        for i in range(2,x+1):
            a , b = b , a + b
        print('hello')
        return a  
start = time.time()
fun_cache(100)
print(time.time()-start)
start = time.time()
fun_cache(100) # used from cache here @lru_cache is only for the first function
print(time.time()-start)
```

Wrapper()

```python
def my_decorator(func):
    @f.wraps(func)
    def wrapper():
        print("Before")
        func()
        print("After")  
    return wrapper
@my_decorator
def greet():
    print("Hi")
print(greet())
```

```python
def print_args(func):
    @f.wraps(func)
    def wrapper(*args,**kwargs):
        print("Calles")
        return func(*args,**kwargs)
    return wrapper
@print_args
def insert(a,b):
    print(a,b)
insert(1,{'name':'shreeram'})
```
